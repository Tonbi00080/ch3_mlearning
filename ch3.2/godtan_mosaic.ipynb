```python
# モザイク関数読み込み
from mosaic import mosaic as mosaic

# カスケードファイルを指定して、検出器を作成
cascade_file = "haarcascade_frontalface_alt.xml"
# カスケード分類器のファイルを読み込む
cascade = cv2.CascadeClassifier(cascade_file)

# 画像を読み込んでグレイスケールに変換
img = cv2.imread("godtan.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 顔認識を実行　顔の位置の配列(x,y,w,h)
# minsize以下の顔は認識されない。顔のサイズが小さいので(20,20)
face_list = cascade.detectMultiScale(img_gray,minSize =(20,20))

# 結果を確認
if len(face_list)  == 0:    quit() # exit
    
# 認識した部分にモザイク
for(x,y,w,h) in face_list:
    # mosaic(画像ファイル,(左上x,左上y,右下x, 右下y), ratio) 
    # ratioの数で分割
    img = mosaic(img, (x,y,x+w, y+h), 10)
    
# 画像を出力する
cv2.imwrite("godtan_mosaic.png", img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
