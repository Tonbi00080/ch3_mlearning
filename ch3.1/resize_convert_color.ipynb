```python
import matplotlib.pyplot as plt
import cv2

# 画像を読みこむ
img = cv2.imread("test.png")
# resize(画像読み込んだ変数,(width,height))
im2 = cv2.resize(img,(200,500))

cv2.imwrite("resized-test.png", im2)

# 画像を表示
# cv2.cvtColor(画像ファイル変数,色空間の変換)
# bgr->rgb
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

# 画像を読みこむ
img = cv2.imread("test.png")
# resize(画像読み込んだ変数,(width,height))
im2 = cv2.resize(img,(200,500))

cv2.imwrite("resized-test.png", im2)

# 画像を表示
# cv2.cvtColor(画像ファイル変数,色空間の変換)
# bgr->rgb
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2HSV))
plt.show()
```
